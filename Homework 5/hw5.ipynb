```python
from pynq import Overlay
axi_gpio_design = Overlay("./hw5.bit")
```

```python
axi_gpio_design.ip_dict
```

```python
cdma0_address = axi_gpio_design.ip_dict['axi_cdma_0']['phys_addr']
print("Physical address of cdma0: 0x" + format(cdma0_address, '02x'))
```

```python
zynq_addr = 0X30000000 #PYNQ address
```

```python
from pynq import MMIO
zynq_ps = MMIO(zynq_addr, 16384)
cdma = MMIO(cdma0_address, 65536)
```

```python
## Open file
fp = open('input.hex', "r")
line = fp.readline()
addr = 0
while line:
    hex_int = int(line, 16)
    
    zynq_ps.write(addr, hex_int)
    line = fp.readline()
    addr = addr+4
fp.close()
```

```python
##write data fram ZYNQ to bram0
cdma.write(0x0, 0x4)#CDMACR
cdma.write(0x18, zynq_addr)#SA
cdma.write(0x20, 0xC0000000)#DA
cdma.write(0x28, 3180)#BTT
```

```python
##write data fram bram1 to ZYNQ
cdma.write(0x0, 0x4)#CDMACR
cdma.write(0x18, 0xC2000000)#SA
cdma.write(0x20, zynq_addr)#DA
cdma.write(0x28, 2704)#BTT
```

```python
fp = open('output.hex', "w")
addr = 0
while addr<=2700:
    hex_int = zynq_ps.read(addr)
    fp.writelines(str('%08x' % hex_int).upper()+"\n")
    print(('%08x' % hex_int).upper())
    addr = addr+4
    
fp.close()
```

```python
#checking the contents of the output and comparing with the golden
f1=open("output.hex","r")
f2=open("golden.hex","r")
i=-1
j=1
for line1 in f1:
    for line2 in f2:
        i=i+1
        if line1!=line2:
            j=0
            print("Different at pixel", i, ":", line1, "expected", line2)
        #else:
            #print("Same at pixel",i)
        break
if j==1:
    print("CONGRATULATIONS!")
else:
    print("OOPS!")
f1.close()
f2.close()
```
